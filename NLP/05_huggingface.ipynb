# Modelo

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/eduardofc/data/main/amazon_sports.csv")
df['review_body'] = df['review_body'].str.replace('[^a-zA-ZñÑáéíóú .,]','', regex=True)
df['review_body'] = df['review_body'].str.lower()
df.head()
```

```python
df = df[df.stars != 2]
df = df[df.stars != 3]
df = df[df.stars != 4]
df['good_product'] = (df.stars>3).astype(int)
df.groupby('good_product').size()
```

```python
X = df['review_body'].to_list()
```

```python
from transformers import AutoTokenizer
from transformers import AutoModel
import torch

# 1) elegimos un modelo en castellano
model_name = "mrm8488/distill-bert-base-spanish-wwm-cased-finetuned-spa-squad2-es"

# 2) preparamos model y tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
X_inputs = tokenizer(X, padding=True, truncation=True, return_tensors='pt')
#outputs = model(**inputs)
```

```python
X_inputs_ids = X_inputs['input_ids']
X_inputs_attention_mask = X_inputs['attention_mask']

X_inputs_ids.shape
```

```python
def forward_pass(x):
  inputs = tokenizer(x, padding=True, truncation=True, return_tensors='pt')
  outputs = model(**inputs)
  last_hidden_state = outputs.last_hidden_state
  with torch.no_grad():
    last_hidden_state = last_hidden_state.cpu().numpy()
  return last_hidden_state[0][1]
```

```python
# Evaluar un batch y pasar de X en X
last_layer_model = df['review_body'].map(forward_pass)
```

```python
# Modelling
pd_last_layer_model = pd.DataFrame(last_layer_model.to_list(), columns=[f'x{i}' for i in range(768)])
pd_last_layer_model.head()
```

```python
from sklearn.model_selection import train_test_split

X = pd_last_layer_model
y = df.good_product

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=2, random_state=99)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=99)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

# AutoModel for Sequence Classification

```python
sentences = [
    "I've been waiting for a HuggingFace course my whole life",
    "I hate this so much!"
]
```

```python
from transformers import AutoModelForSequenceClassification
from transformers import AutoConfig, AutoModel

model_name = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')
print(inputs)
```

```python
outputs = model(**inputs)
print(outputs) #logits tensor
```

```python
config = AutoConfig.from_pretrained(model_name)
config
model.config.id2label
outputs.logits
# 0 - NEGATIVE, 1 - POSITIVE
```

```python
predictions = torch.nn.functional.softmax(outputs.logits)
predictions
```

# All together in a pipeline


```python
from transformers import DistilBertForSequenceClassification, DistilBertTokenizerFast
from transformers import pipeline

model_name = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"

pipe = pipeline(
    task = 'text-classification',
    model = DistilBertForSequenceClassification.from_pretrained(model_name),
    tokenizer = DistilBertTokenizerFast.from_pretrained(model_name)
)
```

```python
pipe(sentences)
```
